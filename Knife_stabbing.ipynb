```python
pred_dir = '/content/drive/My Drive/dataset/runs/detect/train2/labels/'

knife_class_id = 4  # Replace with the correct ID for 'Knife_Weapon'
total_images = 0
knife_detected = 0

if os.path.exists(pred_dir):
    for filename in os.listdir(pred_dir):
        if filename.endswith('.txt'):
            total_images += 1
            with open(os.path.join(pred_dir, filename), 'r') as f:
                for line in f:
                    class_id = int(line.split()[0])
                    if class_id == knife_class_id:
                        knife_detected += 1
                        break

    accuracy = (knife_detected / total_images) * 100
    print(f"Knife Detection Accuracy: {accuracy:.2f}%")
    print(f"Total images: {total_images}, Images with knife detected: {knife_detected}")
else:
    print(f"The directory {pred_dir} does not exist.")
```

```python
# Class IDs for Knife_Weapon and Knife_Deploy (adjust based on your YAML class order)
knife_weapon_class_id = 4  # Replace with the correct class ID for 'Knife_Weapon'
knife_deploy_class_id = 3  # Replace with the correct class ID for 'Knife_Deploy'

# Directory where predictions are saved
pred_dir = '/content/runs/detect/predict/labels/'  # Adjust this path if necessary

total_images = 0
detected_images = 0

if os.path.exists(pred_dir):
    for filename in os.listdir(pred_dir):
        if filename.endswith('.txt'):
            total_images += 1
            with open(os.path.join(pred_dir, filename), 'r') as f:
                for line in f:
                    class_id = int(line.split()[0])
                    if class_id == knife_weapon_class_id or class_id == knife_deploy_class_id:
                        detected_images += 1
                        break  # Stop checking after the first detection

    # Calculate accuracy
    accuracy = (detected_images / total_images) * 100 if total_images > 0 else 0
    print(f"Knife Detection Accuracy: {accuracy:.2f}%")
    print(f"Total images: {total_images}, Images with Knife_Weapon or Knife_Deploy detected: {detected_images}")
else:
    print(f"The directory {pred_dir} does not exist.")
```
